#### Calculation derivative of this function:

(derivate is a measure of: if we slightly change the value x to the right, how the y will respond? Will it be slightly up or down?)

```python
x = 3.0 # x point
h = 0.00000001 # small difference on x axis
y = f(x) # y for x
y_prim = f(x + h) # y after small movement over x axis in positive direction. Derivative would indicate if we move slightly up or down compared to f(x)
y_diff = y_prim - y # how much the function responded on change of x in a positive direction
y_derivative = y_diff / h # rise (y) response must be normalized by run (x) to get a slope
y_derivative
```

# Building micrograd

### Creating structures for expressions

We need this data structure to make expressions which track previous numbers and expressions.

```python
# Concepts introduced in this class Value are explained later in this notebook
class Value:
    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0.0 # this maintains derivative of the output (loss function) in respect to self.data in this Value object
        self._prev = set(_children) # this allows us to connect the previous values
        self._op = _op # this allows for tracking operations that created result from values stored in self._prev
        self.label = label
    
    def __repr__(self):
        return f"Value(data={self.data})"
    
    def __add__(self, other):
        out = Value(self.data + other.data, (self, other), '+')
        return out
    
    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other), '*')
        return out
```

```python
a = Value(2.0, label = 'a')
b = Value(3.0, label = 'b')
c = Value(10.0, label = 'c')
d = a * b; d.label = 'd'
e = d + c; e.label = 'e'
f = Value(-2.0, label='f')
L = e * f ; L.label='L'
L
```

We need those functions to visualize the network

```python
from graphviz import Digraph

def trace(root):
    nodes, edges = set(), set()
    def build(v):
        if v not in nodes:
            nodes.add(v)
            for child in v._prev:
                edges.add((child, v))
                build(child)
    build(root)
    return nodes, edges

def draw_dot(root):
    dot = Digraph(format='SVG', graph_attr={'rankdir': 'LR'}) #, node_attr={'rankdir': 'TB'})
    
    nodes, edges = trace(root)    
    for n in nodes:
        dot.node(name=str(id(n)), label = "{ %s | data %.4f | grad %.4f }" % (n.label, n.data, n.grad), shape='record')
        if n._op:
            dot.node(name=str(id(n)) + n._op, label=n._op)
            dot.edge(str(id(n)) + n._op, str(id(n)))
    
    for n1, n2 in edges:
        dot.edge(str(id(n1)), str(id(n2)) + n2._op)
    
    return dot

draw_dot(L)
```

#### Backpropagation "by hand"

Backpropagation is the process of checking impact of each node (a,b,c,d,e,f) on the output - last node (L).  
This impact is measured by derivative. Definition of derivative is: (f(x+h) - f(x))/h  
So impact of f on output is the derivative of L with respect to f:  
dL / df  
that is (from definition (f(x+h) - f(x))/h ):  
L = f * e  
((f+h * e) - f * e)/h  
(fe + he - fe)/h  
he/h  
e
(impact of f to L is e)  
because of that we put f.grad = e.data and e.grad = f.data 

```python
f.grad = e.data; e.grad = f.data
```

Let's check if that's true with recomputing the values with the example function:

```python
def check_grad(node):
    a = Value(2.0, label = 'a')
    b = Value(3.0, label = 'b')
    c = Value(10.0, label = 'c')
    d = a * b
    e = d + c
    f = Value(-2.0, label='f')
    L1 = e * f
    # change:
    h = Value(0.001)
    a = Value(2.0) + h if node == 'a' else Value(2.0)
    b = Value(3.0) + h if node == 'b' else Value(3.0)
    c = Value(10.0) + h if node == 'c' else Value(10.0)
    d = (a * b) + h  if node == 'd' else a * b
    e = d + c + h if node == 'e' else d + c
    f = Value(-2.0) + h if node == 'f' else Value(-2.0)
    L2 = e * f + h if node == 'L' else e * f
    return (L2.data-L1.data)/h.data # this is derivative of L2 in respect to changed variable
```

All right, we got function now let's check if indeed derivative of L with respect to f is e:

```python
check_grad('f') # it's 15.9999999 so it is 16, which is equal to e.data
```

Now let's check if derivative of L with respect to e is the value of f:

```python
check_grad('e')
```

YES! So both computations are correct.  

Now we know that impact of e and f to L.  
Let's got further left. We got c, and we need to measure it's impact on the L.  
But c has imidiate impact on e, and e has impact on L.  

Here we use CHAIN RULE: we multiply intermediate derivatives to measure the impact of some variable to the output.  
[Chain rule on wikipedia](https://en.wikipedia.org/wiki/Chain_rule)  
So to measure impact of c to L we need do:  
(de / dc) * (dL / de)  

What is the impact of c to e?  
e = c + d  
(definition: (f(x+h)-f(x))/h )  
((c + d + h) - (c + d)) / h  
(c + d + h - c - d) / h  
h/h  
1  

So derivative is 1. Now let's compute derivative of dL / dc:  
(de / dc) * (dL / de)  
1 * -2.0 = -2.0  
Impact of c to L is -2.0. It is the same as impact of e.  
That shows that addition only passes the gradient, but does not change anything.  

Let's check if computation is correct!

```python
check_grad('c')
```

Yep, it's -2, which is the same output (if we round) as check_grad('e')

```python
round(check_grad('c'), 1) == round(check_grad('e'), 1)
```

And that's how we do back propagation, we go through each node and measure impact of each node (later called weight) on the output.  

Let's go with a and b (d is the same as c, because it's addition, so we skip that).  
First a. To compute dL / da we do: (dd / da) * (de / dd) * (dL / de)  
We computed dL / de (will be stored in e.grad)  
also computed de / dd (will be stored in d.grad)  
Now let's compute dd / da:
(def: (f(x+h)-f(x))/h)  
d = a * b  
((a+h * b) - (a * b)) / h  
(ab + hb - ab) / h  
hb / h  
b
And b = 3  
We know that d.grad (which contains (de / dd) * (dL / de)) is -2.0 (same as c.grad)  
so we compute dL / da (a.grad) by:  
3 * -2.0 = (b.data * d.grad) = -6.0  

Similar thing will be done for b (so we skip it), that is b.grad = a.data * d.grad
Let's test it:

```python
check_grad('a')
```

```python
round(check_grad('b'), 1)
```

Let's update our graph by impacts on L (derivatives: dL/dnode):

```python
for node in [a,b,c,d,e,f,L]:
    node.grad = check_grad(node.label)
draw_dot(L)
```

(note that L.grad = 1, because impact of L to L is 1 :)

### Backpropagation in neurons

Neurons are mathematically modeled biological neurons. They are simple:  
  
https://cs231n.github.io/assets/nn1/neuron_model.jpeg

Each input data x is multiplied by weight w.   
All those are added (plus bias b) in the cell body.  
Output is passed to activation function f. Result from activation function is passed to another neuron (in another layer).  
Here is example of activation function tanh:  

```python
plt.plot(np.arange(-5.0, 5.0, 0.25), np.tanh(np.arange(-5.0, 5.0, 0.25))); plt.grid()
```

note that output from this activation function is between -1.0 and 1.0. For this function it flattens for x less than -2 and higher than 2.  

Let's add tanh() function to the Vaule class:

```python
# Concepts introduced in this class Value are explained later in this notebook
class Value:
    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0.0 # this maintains derivative of the output (loss function) in respect to self.data in this Value object
        self._prev = set(_children) # this allows us to connect the previous values
        self._op = _op # this allows for tracking operations that created result from values stored in self._prev
        self.label = label
    
    def __repr__(self):
        return f"Value(data={self.data})"
    
    def __add__(self, other):
        out = Value(self.data + other.data, (self, other), '+')
        return out
    
    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other), '*')
        return out
    
    def tanh(self): # this is activation function that is used in the neuron body (explained below). Definition from https://en.wikipedia.org/wiki/Hyperbolic_functions
        x = self.data
        t = (math.exp(2*x)-1)/(math.exp(2*x)+1)
        out = Value(t, (self, ), 'tanh')
        return out
    
```

Let's create a simple neuron:

```python
def simple_neuron(x1, x2, w1, w2, b):
    # inputs (data):
    x1 = Value(x1, label='x1')
    x2 = Value(x2, label='x2')
    # weights:
    w1 = Value(w1, label='w1')
    w2 = Value(w2, label='w2')
    # bias:
    b = Value(b, label='b')
    # we compute cell body: x1*w1 + x2*w2 + b
    # but we need to get all the intermediate nodes
    x1w1 = x1*w1; x1w1.label='x1w1'
    x2w2 = x2*w2; x2w2.label='x2w2'
    x1w1x2w2 = x1w1 + x2w2; x1w1x2w2.label='x1*w1 + x2*w2'
    n = x1w1x2w2 + b; n.label='n'
    # here we use activation function
    o = n.tanh(); o.label='o'
    # return all intermediate nodes
    return {'x1': x1, 'x2': x2, 'w1': w1, 'w2': w2, 'b': b, 'x1w1':x1w1, 'x2w2':x2w2, 'x1w1x2w2':x1w1x2w2, 'n':n, 'o':o}
    
neuron_output = simple_neuron(x1=2.0, x2=0.0, w1=-3.0, w2=1.0, b=8.0)
neuron_output['o']
```

The function activation function tanh is squashing the output (2.0) to the -1,1 range, in this case 0.964  

Now let's generate new neron we need to backpropagate in this neuron.

```python
neuron_output = simple_neuron(x1=2.0, x2=0.0, w1=-3.0, w2=1.0, b=6.8813735870195432)
draw_dot(neuron_output['o'])
```

To do backpropagation on tanh() function we can go to [wikipedia](https://en.wikipedia.org/wiki/Hyperbolic_functions) and check what is the derivative of tanh().  
Definition looks like this:  
dtanh(x) / dx = 1 - tanh(x)^2  

Now we know how to do derivative of do / dn (i.e. n.grad):  
(because o = tanh(n))  
do / dn = 1 - o^2

```python
neuron_output['o'].grad = 1 # it must be one, because derivative of o in respect to o is always 1
neuron_output['n'].grad = round(1 - neuron_output['o'].data**2, 1)
neuron_output['n'].grad
```

So it looks like the derivative do / dn is 0.5.

Next are w2 and x2. This is simple product so grad will be:

```python
neuron_output['w2'].grad = neuron_output['x2'].data * neuron_output['x2w2'].grad
neuron_output['x2'].grad = neuron_output['w2'].data * neuron_output['x2w2'].grad
```

#### Automating backpropagation 

*Here we added _backward() functionality to the Value class.*

```python
class Value:
    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0.0 # this maintains derivative of the output (loss function) in respect to self.data in this Value object
        self._backward = lambda: None # this is function that chains output gradient with input gradients
        self._prev = set(_children) # this allows us to connect the previous values
        self._op = _op # this allows for tracking operations that created result from values stored in self._prev
        self.label = label
    
    def __repr__(self):
        return f"Value(data={self.data})"
    
    def __add__(self, other):
        out = Value(self.data + other.data, (self, other), '+')
        
        def _backward(): # we define how the gradient is computed
            self.grad = 1.0 * out.grad # since it is addition, the gradient from output is just copied to this node, and the other
            other.grad = 1.0 * out.grad # we multiply by one to show the derivative
        
        out._backward = _backward
        
        return out
    
    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other), '*')
        
        def _backward():
            self.grad = other.data * out.grad # local derivative for multiplication is value of the other. Since it is chain rule, we multiply it by out gradient.
            other.grad = self.data * out.grad
            
        out._backward = _backward
        
        return out
    
    def tanh(self): # this is activation function that is used in the neuron body (explained below). Definition from https://en.wikipedia.org/wiki/Hyperbolic_functions
        x = self.data
        t = (math.exp(2*x)-1)/(math.exp(2*x)+1)
        out = Value(t, (self, ), 'tanh')
        
        def _backward():
            self.grad = (1 - out.data**2) * out.grad # this is local derivative of tanh. We multiply it by output gradient, because of the chain rule
        
        out._backward = _backward
        
        return out
    
```

Let's assign gradient to "o" and to "n". To "o" we assign 1.0 by hand. But to assign n.grad we cal o._backward

```python
neuron_output = simple_neuron(x1=2.0, x2=0.0, w1=-3.0, w2=1.0, b=6.8813735870195432)
neuron_output['o'].grad = 1.0 # we need to set it to 1.0 because by default Value.grad is 0
neuron_output['o']._backward() # here we assign value to n.grad
```

```python
draw_dot(neuron_output['o'])
```

The rest of grads:

```python
neuron_output['n']._backward()
neuron_output['x1w1x2w2']._backward()
neuron_output['x1w1']._backward()
neuron_output['x2w2']._backward()
```

```python
draw_dot(neuron_output['o'])
```

Great, it works! Note that if we call _backward() on b, w2, x2, w1, x1 nothing would happen, because those have no function

This is time to create functionality that allows to backpropagate, after the neuron did all computing (so when we create data for each node, then we can backpropagate)

```python
class Value:
    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0.0 # this maintains derivative of the output (loss function) in respect to self.data in this Value object
        self._backward = lambda: None # this is function that chains output gradient with input gradients. By default it's empty
        self._prev = set(_children) # this allows us to connect the previous values
        self._op = _op # this allows for tracking operations that created result from values stored in self._prev
        self.label = label
    
    def __repr__(self):
        return f"Value(data={self.data})"
    
    def __add__(self, other):
        other = other if isinstance(other, Value) else Value(other) # allows to do: Value(data=1) + 1
        out = Value(self.data + other.data, (self, other), '+')
        
        def _backward():
            '''
            We define how the gradient is computed.
            Since it is addition, the gradient from output is just copied to this node, and the other.
            We need to accumulate gradient if we use variable more than once, that's why +=
            '''
            self.grad += 1.0 * out.grad
            other.grad += 1.0 * out.grad # we multiply by one to show the derivative
        
        out._backward = _backward
        
        return out
    
    def __mul__(self, other):
        other = other if isinstance(other, Value) else Value(other)
        out = Value(self.data * other.data, (self, other), '*')
        
        def _backward():
            self.grad += other.data * out.grad # local derivative for multiplication is value of the other. Since it is chain rule, we multiply it by out gradient.
            other.grad += self.data * out.grad
            
        out._backward = _backward
        
        return out
    
    def __rmul(self,other): # fallback for int(2) * Value(2)
        return self * other
    
    def tanh(self): # this is activation function that is used in the neuron body (explained below). Definition from https://en.wikipedia.org/wiki/Hyperbolic_functions
        x = self.data
        t = (math.exp(2*x)-1)/(math.exp(2*x)+1)
        out = Value(t, (self, ), 'tanh')
        
        def _backward():
            self.grad += (1 - out.data**2) * out.grad # this is local derivative of tanh. We multiply it by output gradient, because of the chain rule
        
        out._backward = _backward
        
        return out
    
    def backward(self):
        # here is the topological graph implementation
        # it will addpend all nodes to topo[], and including itself
        topo = []
        visited = set()
        def build_topo(v):
            if v not in visited:
                visited.add(v)
                for child in v._prev:
                    build_topo(child)
                topo.append(v)
        
        build_topo(self)
        self.grad = 1.0 # initialize output gradient to 1.0 (if it would stay 0.0 everywhere would be zero because of chain rule)
        # running backpropagation
        for node in reversed(topo):
            node._backward()
            
    
```

Let's see if it works:

```python
neuron_output = simple_neuron(x1=2.0, x2=0.0, w1=-3.0, w2=1.0, b=6.8813735870195432)
neuron_output['o'].backward() # here we do backpropagation just on the output node
draw_dot(neuron_output['o'])
```

### Creating subfunctions of tanh: exp and division. Also additional functions

```python
class Value:
    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0.0 # this maintains derivative of the output (loss function) in respect to self.data in this Value object
        self._backward = lambda: None # this is function that chains output gradient with input gradients. By default it's empty
        self._prev = set(_children) # this allows us to connect the previous values
        self._op = _op # this allows for tracking operations that created result from values stored in self._prev
        self.label = label
    
    def __repr__(self):
        return f"Value(data={self.data})"
    
    def __add__(self, other):
        other = other if isinstance(other, Value) else Value(other) # allows to do: Value(data=1) + 1
        out = Value(self.data + other.data, (self, other), '+')
        
        def _backward():
            '''
            We define how the gradient is computed.
            Since it is addition, the gradient from output is just copied to this node, and the other.
            We need to accumulate gradient if we use variable more than once, that's why +=
            '''
            self.grad += 1.0 * out.grad
            other.grad += 1.0 * out.grad # we multiply by one to show the derivative
        
        out._backward = _backward
        return out
    
    def __mul__(self, other):
        other = other if isinstance(other, Value) else Value(other)
        out = Value(self.data * other.data, (self, other), '*')
        
        def _backward():
            self.grad += other.data * out.grad # local derivative for multiplication is value of the other. Since it is chain rule, we multiply it by out gradient.
            other.grad += self.data * out.grad
            
        out._backward = _backward
        return out
    
    def __pow__(self, other):
        assert isinstance(other, (int, float)), "only support int or float power for now"
        out = Value(self.data**other, (self, ), f'**{other}')
        
        def _backward():
            self.grad += other * (self.data ** (other - 1)) * out.grad # derivative of x**n is nx**(x-1). Multiply by grad (chain rule)
        
        out._backward = _backward
        return out
            
    
    def __rmul__(self, other): # fallback for int(2) * Value(2)
        return self * other # here we have implemented our definition of multiplication, so we don't need to wrap it in Value and define bakcward function
    
    def __truediv__(self, other): # division
        return self * other**-1 # self * (1/other) = self / other
    
    def __neg__(self):
        return self * -1
    
    def __sub__(self, other):
        return self + (-other)
    
    def tanh(self): # this is activation function that is used in the neuron body (explained below). Definition from https://en.wikipedia.org/wiki/Hyperbolic_functions
        x = self.data
        t = (math.exp(2*x)-1)/(math.exp(2*x)+1)
        out = Value(t, (self, ), 'tanh')
        
        def _backward():
            self.grad += (1 - out.data**2) * out.grad # this is local derivative of tanh. We multiply it by output gradient, because of the chain rule
        
        out._backward = _backward
        return out
    
    def exp(self):
        x = self.data
        out = Value(math.exp(x), (self,), label='exp')
        
        def _backward():
            self.grad += out.data * out.grad # derivative of e^x is e^x, we have e^x in out.data so we mutiply it by out.grad (becasue of chain rule)
        
        out._backward = _backward
        return out
    
    def backward(self):
        # here is the topological graph implementation
        # it will addpend all nodes to topo[], and including itself
        topo = []
        visited = set()
        def build_topo(v):
            if v not in visited:
                visited.add(v)
                for child in v._prev:
                    build_topo(child)
                topo.append(v)
        
        build_topo(self)
        self.grad = 1.0 # initialize output gradient to 1.0 (if it would stay 0.0 everywhere would be zero because of chain rule)
        # running backpropagation
        for node in reversed(topo):
            node._backward()
            
    
```

Let's break tanh() function into more simple functions:

```python
def simple_neuron(x1, x2, w1, w2, b):
    # inputs (data):
    x1 = Value(x1, label='x1')
    x2 = Value(x2, label='x2')
    # weights:
    w1 = Value(w1, label='w1')
    w2 = Value(w2, label='w2')
    # bias:
    b = Value(b, label='b')
    # we compute cell body: x1*w1 + x2*w2 + b
    # but we need to get all the intermediate nodes
    x1w1 = x1*w1; x1w1.label='x1w1'
    x2w2 = x2*w2; x2w2.label='x2w2'
    x1w1x2w2 = x1w1 + x2w2; x1w1x2w2.label='x1*w1 + x2*w2'
    n = x1w1x2w2 + b; n.label='n'
    # replacing tanh() with atomic functions:
    e = (2*n).exp()
    o = (e-1)/(e+1); o.label='o'
    # return all intermediate nodes
    return {'x1': x1, 'x2': x2, 'w1': w1, 'w2': w2, 'b': b, 'x1w1':x1w1, 'x2w2':x2w2, 'x1w1x2w2':x1w1x2w2, 'n':n, 'e':e, 'o':o}

neuron_output = simple_neuron(x1=2.0, x2=0.0, w1=-3.0, w2=1.0, b=6.8813735870195432)
neuron_output['o'].backward()
draw_dot(neuron_output['o'])
```

### Let's do the same thing with pyTorch 

```python
import torch
import random
```

This is the same thing in pytorch.  
Our Value object takes scalar value (just float number). But Tensor from pytorch takes tensor, that is multidimensional array. Here we pass just scalar in the array, and we cast to the double type. By default nodes do not require gradient, so we need to set it to True.

```python
x1 = torch.Tensor([2.0]).double() ;  x1.requires_grad = True
x2 = torch.Tensor([0.0]).double() ;  x2.requires_grad = True
w1 = torch.Tensor([-3.0]).double() ; w1.requires_grad = True
w2 = torch.Tensor([1.0]).double() ;  w2.requires_grad = True
b = torch.Tensor([6.8813735870195432]).double() ;  b.requires_grad = True
n = x1*w1 + x2*w2 + b
o = torch.tanh(n)

print(o) # this is a tensor
print(o.item()) # we can return also single element
o.backward()

print('----')
print('x2', x2.grad.item())
print('w2', w2.grad.item())
print('x1', x1.grad.item())
print('w1', w1.grad.item())
```

### Creation of neuron, layer and multi-layer perceptron classes

The above structure is Multi Layer Perceptron (MLP).  
It consists of input layer with data, hidden layers with neuron, and with output layer.  
We need to create three classes: Neuron, Layer, and MLP.

```python
class Neuron:
    def __init__(self, nin): # nin is number of inputs
        self.w = [Value(random.uniform(-1, 1)) for _ in range(nin)]
        self.b = Value(random.uniform(-1, 1))
        
    def __call__(self, x):
        # this is a default call function, so we can do: n = Neuron(2); n([1,2])
        # w * x + b
        n = sum([wi * xi for wi, xi in zip(self.w, x)], start=self.b) # we start from bias
        out = n.tanh()
        return out
    
    def parameters(self):
        return self.w + [self.b]

class Layer:
    def __init__(self, nin, nout): # nin is for Neuron init. nout is the number of output neurons in that layer
        self.neurons = [Neuron(nin) for _ in range(nout)]

    def __call__(self,x):
        outs = [n(x) for n in self.neurons]
        return outs[0] if len(outs) == 1 else outs
    
    def parameters(self):
        return [param for neuron in self.neurons for param in neuron.parameters()]

class MLP:
    def __init__(self, nin, nouts): # nounts is array of numbers which represent Neurons in each layer
        sizes = [nin] + nouts # we want to merge it, because the first is number of inputs for the first layer, and then each number of outputs is input for next layer
        self.layers = [Layer(sizes[i], sizes[i+1]) for i in range(len(nouts))]
    
    def __call__(self, x):
        for layer in self.layers:
            x = layer(x) # output values from each layer are inputs to next layer
        return x # return output from last layer
    
    def parameters(self):
        return [param for layer in self.layers for param in layer.parameters()]
    
```

x - his is input data for input layer  
\[4, 4, 1\] - first layer: 4 neurons, second layer: 4 neurons, last layer: 1 output neuron

```python
x = [2.0, 3.0, -1.0]
mlp = MLP(3, [4, 4, 1])
mlp(x)
```

### Nerual net training 

Let's try to train our net.  
Ok, so now we see what our network thinks about the data.  

```python
# this is our data. Each array has 3 input data data
xs = [
    [2.0, 3.0, -1.0], # this should give ys[0] = 1.0
    [3.0, -1.0, 0.5], # this should give ys[1] = -1.0
    [0.5, 1.0, 1.0],  # this should give ys[2] = -1.0
    [1.0, 1.0, -1.0]  # this should give ys[3] = 1.0
]
ys = [1.0, -1.0, -1.0, 1.0] # those are desired targets
mlp = MLP(3, [4, 4, 1])  # 3 input data, 2 hidden layers with 4 neurons each, last layer with 1 output data
y_pred = [mlp(x) for x in xs]
y_pred
```

How do we adjust the weights to predict better?  
The trick is, to calculate one number that would measure the total performance of our neural net.  
We call this single number "the loss".  
Here we'll implement mean squared error loss:

```python
[(yout - y_ground_truth)**2 for y_ground_truth, yout in zip(ys, y_pred)]
```

```python
loss = sum([(yout - y_ground_truth)**2 for y_ground_truth, yout in zip(ys, y_pred)], start=Value(0.0))
loss
```

We want to minimize the loss, because if loss is 0 then we predicted perfectly.  

Ok, we got loss, which shows us how "bad" our neural net is.  
To now improve it we need to adjust weights - up or down - depending on what was prediction and what is impact of each weight for loss.  
For example, this weight:

```python
mlp.layers[0].neurons[0].w[0]
```

currently does not have computed gradient, which is impact for loss:

```python
mlp.layers[0].neurons[0].w[0].grad
```

but we can easily compute it:

```python
loss.backward()
```

```python
mlp.layers[0].neurons[0].w[0].grad
```

So if gradient of this wieght is negative, then we need to increase the value of that weight to make loss smaller.  
And if gradient of this wieght is positive, then we need to decrease the value of that weight to make loss smaller.
